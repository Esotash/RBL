```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('NSE50.csv')

# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculate Simple Moving Average (SMA)
def calculate_sma(data, window):
    return data['Close'].rolling(window=window).mean()

# Calculate Exponential Moving Average (EMA)
def calculate_ema(data, window):
    return data['Close'].ewm(span=window, adjust=False).mean()

# Define window sizes for SMA and EMA
windows = [20, 50, 100, 150, 200]

# Calculate SMA and EMA for each window size and add to DataFrame
for window in windows:
    data[f'SMA_{window}'] = calculate_sma(data, window)
    data[f'EMA_{window}'] = calculate_ema(data, window)

# Display the updated DataFrame with SMAs and EMAs
print(data[['Date', 'Close'] + [f'SMA_{window}' for window in windows] + [f'EMA_{window}' for window in windows]].head())

# Optional: Plotting
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')

# Plot SMA and EMA
for window in windows:
    plt.plot(data['Date'], data[f'SMA_{window}'], label=f'{window}-Day SMA', linestyle='--')
    plt.plot(data['Date'], data[f'EMA_{window}'], label=f'{window}-Day EMA', linestyle=':')

# Save the DataFrame with SMA and EMA data to a new CSV file
data.to_csv('NSE5001.csv', index=False)

plt.title('Nifty 50 Close Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()  # Calculate the difference between consecutive close prices
    gain = np.where(delta > 0, delta, 0)  # Take only the positive differences (gains)
    loss = np.where(delta < 0, -delta, 0)  # Take only the negative differences (losses)

    avg_gain = pd.Series(gain).rolling(window=window, min_periods=1).mean()  # Rolling average of gains
    avg_loss = pd.Series(loss).rolling(window=window, min_periods=1).mean()  # Rolling average of losses

    rs = avg_gain / avg_loss  # Relative Strength (RS)
    rsi = 100 - (100 / (1 + rs))  # RSI formula

    data['RSI'] = rsi  # Add RSI to the dataframe
    return data

# Apply RSI to the Nifty 50 OHLC data
nifty_data = pd.read_csv('NSE50.csv')  # Load data (assumed already saved)
nifty_data = calculate_rsi(nifty_data)

# Display the data with RSI
print(nifty_data[['Date','Open','High','Low','Close','RSI']].tail())


```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load your dataset
nifty_data = pd.read_csv('NSE5001.csv')  # Adjust the path to your dataset

# Calculate Moving Averages
nifty_data['MA20'] = nifty_data['Close'].rolling(window=20).mean()
nifty_data['MA50'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['MA100'] = nifty_data['Close'].rolling(window=100).mean()
nifty_data['MA150'] = nifty_data['Close'].rolling(window=150).mean()
nifty_data['MA200'] = nifty_data['Close'].rolling(window=200).mean()

# Ensure that 'RSI' and 'MACD' are calculated and included in nifty_data

# Select features and target
features = nifty_data[['RSI', 'MACD', 'MA20', 'MA50', 'MA100', 'MA150', 'MA200']]
target = nifty_data['Close']

# Normalize the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Convert target to numpy array
target_scaled = target.values

# Define the time step
time_step = 60  # Number of previous days to consider for prediction

# Create the datasets
X, y = [], []
for i in range(time_step, len(features_scaled)):
    X.append(features_scaled[i-time_step:i])
    y.append(target_scaled[i])

X, y = np.array(X), np.array(y)

# Split into training and testing sets (80% training, 20% testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Check the shapes of your datasets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt

# Load your dataset
nifty_data = pd.read_csv('NSE5001.csv')

# Calculate Moving Averages
nifty_data['MA20'] = nifty_data['Close'].rolling(window=20).mean()
nifty_data['MA50'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['MA100'] = nifty_data['Close'].rolling(window=100).mean()
nifty_data['MA150'] = nifty_data['Close'].rolling(window=150).mean()
nifty_data['MA200'] = nifty_data['Close'].rolling(window=200).mean()

# Calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

nifty_data['RSI'] = calculate_rsi(nifty_data)

# Calculate MACD
def calculate_macd(data):
    exp1 = data['Close'].ewm(span=12, adjust=False).mean()
    exp2 = data['Close'].ewm(span=26, adjust=False).mean()
    macd = exp1 - exp2
    return macd

nifty_data['MACD'] = calculate_macd(nifty_data)

# Check for NaN values after calculations
print(nifty_data.isnull().sum())

# Drop rows with NaN values
nifty_data.dropna(inplace=True)

# Select features and target after handling NaN values
features = nifty_data[['RSI', 'MACD', 'MA20', 'MA50', 'MA100', 'MA150', 'MA200']]
target = nifty_data['Close']

# Normalize the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Convert target to numpy array
target_scaled = target.values

# Define the time step
time_step = 60

# Create the datasets
X, y = [], []
for i in range(time_step, len(features_scaled)):
    X.append(features_scaled[i - time_step:i])
    y.append(target_scaled[i])

X, y = np.array(X), np.array(y)

# Split into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Ensure the correct data type
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=10)

# Plot the training loss
plt.plot(model.history.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

```python
predictions = model.predict(X_test)
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load your dataset
nifty_data = pd.read_csv('NSE5001.csv')  # Adjust the path to your dataset

# Calculate Moving Averages
nifty_data['MA20'] = nifty_data['Close'].rolling(window=20).mean()
nifty_data['MA50'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['MA100'] = nifty_data['Close'].rolling(window=100).mean()
nifty_data['MA150'] = nifty_data['Close'].rolling(window=150).mean()
nifty_data['MA200'] = nifty_data['Close'].rolling(window=200).mean()

# Drop rows with NaN values (you can also impute missing values if needed)
nifty_data.dropna(inplace=True)

# Prepare features and target variable
features = nifty_data[['Close', 'MA20', 'MA50', 'MA100', 'MA150', 'MA200']].values
target = nifty_data['Close'].values

# Scale the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Create sequences for LSTM input
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length, 0]  # Target is the Close price
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Define the time step
time_step = 60  # Number of previous days to consider for prediction
X, y = create_sequences(features_scaled, time_step)

# Split into training and testing sets (80% training, 20% testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the LSTM model creation function
def create_model(units=50, dropout=0.2):
    model = Sequential()
    model.add(LSTM(units, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(dropout))
    model.add(LSTM(units, return_sequences=False))
    model.add(Dropout(dropout))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Hyperparameters to test
units_list = [50, 100]
dropout_list = [0.2, 0.3]
batch_size_list = [32, 64]
epochs_list = [10, 20]

best_score = float('inf')
best_params = {}

# Manual grid search
for units in units_list:
    for dropout in dropout_list:
        for batch_size in batch_size_list:
            for epochs in epochs_list:
                print(f'Testing: units={units}, dropout={dropout}, batch_size={batch_size}, epochs={epochs}')

                # Create and train the model
                model = create_model(units=units, dropout=dropout)
                model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=0)

                # Evaluate the model on test data
                test_loss = model.evaluate(X_test, y_test, verbose=0)
                print(f'Test Loss: {test_loss}')

                # Update the best model if the current one performs better
                if test_loss < best_score:
                    best_score = test_loss
                    best_params = {'units': units, 'dropout': dropout, 'batch_size': batch_size, 'epochs': epochs}

print("Best score: ", best_score)
print("Best hyperparameters: ", best_params)
```
