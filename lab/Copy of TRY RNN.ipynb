```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('/content/NSE50.csv')

# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculate Simple Moving Average (SMA)
def calculate_sma(data, window):
    return data['Close'].rolling(window=window).mean()

# Calculate Exponential Moving Average (EMA)
def calculate_ema(data, window):
    return data['Close'].ewm(span=window, adjust=False).mean()

# Define window sizes for SMA and EMA
windows = [20, 50, 100, 150, 200]

# Calculate SMA and EMA for each window size and add to DataFrame
for window in windows:
    data[f'SMA_{window}'] = calculate_sma(data, window)
    data[f'EMA_{window}'] = calculate_ema(data, window)

# Display the updated DataFrame with SMAs and EMAs
print(data[['Date', 'Close'] + [f'SMA_{window}' for window in windows] + [f'EMA_{window}' for window in windows]].head())

# Optional: Plotting
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')

# Plot SMA and EMA
for window in windows:
    plt.plot(data['Date'], data[f'SMA_{window}'], label=f'{window}-Day SMA', linestyle='--')
    plt.plot(data['Date'], data[f'EMA_{window}'], label=f'{window}-Day EMA', linestyle=':')

# Save the DataFrame with SMA and EMA data to a new CSV file
data.to_csv('NSE5001.csv', index=False)

plt.title('Nifty 50 Close Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()  # Calculate the difference between consecutive close prices
    gain = np.where(delta > 0, delta, 0)  # Take only the positive differences (gains)
    loss = np.where(delta < 0, -delta, 0)  # Take only the negative differences (losses)

    avg_gain = pd.Series(gain).rolling(window=window, min_periods=1).mean()  # Rolling average of gains
    avg_loss = pd.Series(loss).rolling(window=window, min_periods=1).mean()  # Rolling average of losses

    rs = avg_gain / avg_loss  # Relative Strength (RS)
    rsi = 100 - (100 / (1 + rs))  # RSI formula

    data['RSI'] = rsi  # Add RSI to the dataframe
    return data

# Apply RSI to the Nifty 50 OHLC data
nifty_data = pd.read_csv('/content/NSE50.csv')  # Load data (assumed already saved)
nifty_data = calculate_rsi(nifty_data)

# Display the data with RSI
print(nifty_data[['Date','Open','High','Low','Close','RSI']].tail())


```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load your dataset
nifty_data = pd.read_csv('/content/NSE5001.csv')  # Adjust the path to your dataset

# Calculate Moving Averages
nifty_data['MA20'] = nifty_data['Close'].rolling(window=20).mean()
nifty_data['MA50'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['MA100'] = nifty_data['Close'].rolling(window=100).mean()
nifty_data['MA150'] = nifty_data['Close'].rolling(window=150).mean()
nifty_data['MA200'] = nifty_data['Close'].rolling(window=200).mean()

# Ensure that 'RSI' and 'MACD' are calculated and included in nifty_data

# Select features and target
features = nifty_data[['RSI', 'MACD', 'MA20', 'MA50', 'MA100', 'MA150', 'MA200']]
target = nifty_data['Close']

# Normalize the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Convert target to numpy array
target_scaled = target.values

# Define the time step
time_step = 60  # Number of previous days to consider for prediction

# Create the datasets
X, y = [], []
for i in range(time_step, len(features_scaled)):
    X.append(features_scaled[i-time_step:i])
    y.append(target_scaled[i])

X, y = np.array(X), np.array(y)

# Split into training and testing sets (80% training, 20% testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Check the shapes of your datasets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the LSTM model creation function
def create_model(units=50, dropout=0.2):
    model = Sequential()
    model.add(LSTM(units, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(dropout))
    model.add(LSTM(units, return_sequences=False))
    model.add(Dropout(dropout))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# Hyperparameters to test
units_list = [50, 100]
dropout_list = [0.2, 0.3]
batch_size_list = [32, 64]
epochs_list = [10, 20]

best_score = float('inf')
best_params = {}

# Manual grid search
for units in units_list:
    for dropout in dropout_list:
        for batch_size in batch_size_list:
            for epochs in epochs_list:
                print(f'Testing: units={units}, dropout={dropout}, batch_size={batch_size}, epochs={epochs}')

                # Create and train the model
                model = create_model(units=units, dropout=dropout)
                model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=0)

                # Evaluate the model on test data
                test_loss = model.evaluate(X_test, y_test, verbose=0)
                print(f'Test Loss: {test_loss}')

                # Update the best model if the current one performs better
                if test_loss < best_score:
                    best_score = test_loss
                    best_params = {'units': units, 'dropout': dropout, 'batch_size': batch_size, 'epochs': epochs}

print("Best score: ", best_score)
print("Best hyperparameters: ", best_params)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Select the relevant columns for feature engineering
features = nse5001_data[['Open', 'High', 'Low', 'Close', 'Volume',
                           'EMA_20', 'EMA_50', 'EMA_100',
                           'EMA_150', 'EMA_200', 'RSI', 'MACD']]
target = nse5001_data['Close']  # Assuming you want to predict the 'Close' price

# Normalize the feature data
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)

# Prepare the dataset for the RNN
def create_dataset(data, target, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), :])
        y.append(target[i + time_step])
    return np.array(X), np.array(y)

# Set the time step
time_step = 60  # Using the last 60 days for prediction
X, y = create_dataset(scaled_features, target.values, time_step)

# Split the data into training and testing sets
split_ratio = 0.8
train_size = int(len(X) * split_ratio)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout

def create_rnn_model(units=50, dropout=0.2):
    model = Sequential()
    model.add(SimpleRNN(units, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(dropout))
    model.add(SimpleRNN(units, return_sequences=False))
    model.add(Dropout(dropout))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Train the model
model = create_rnn_model(units=50, dropout=0.2)
model.fit(X_train, y_train, epochs=20, batch_size=32)
```

```python
# Assuming rnn_predictions is the output from your RNN model
rnn_predictions = model.predict(X_test)

# Step 1: Reshape the predictions
rnn_predictions_reshaped = rnn_predictions.reshape(-1, 1)  # Reshape to (n_samples, 1)

# Step 2: Prepare for inverse transformation
# Assuming the scaler was fitted on the first 6 features, which includes the target 'Close'
# You want to concatenate zeros for the first 5 features
# Concatenate the reshaped predictions with zeros
zeros_for_scaler = np.zeros((rnn_predictions.shape[0], 5))  # Create an array of zeros for 5 features
combined_data = np.concatenate((zeros_for_scaler, rnn_predictions_reshaped), axis=1)  # Combine them

# Step 3: Inverse transform the predictions
rnn_predictions_actual = scaler.inverse_transform(combined_data)[:, -1]  # Get the last column (predicted 'Close')

# Assuming y_test is the actual values you want to compare against
# Reshape y_test for inverse transformation as well
y_test_actual_reshaped = y_test.reshape(-1, 1)
y_test_zeros = np.zeros((y_test_actual_reshaped.shape[0], 5))  # Prepare zeros for inverse scaling
y_test_combined = np.concatenate((y_test_zeros, y_test_actual_reshaped), axis=1)  # Combine with zeros

# Inverse transform actual values
y_test_actual = scaler.inverse_transform(y_test_combined)[:, -1]  # Get the last column (actual 'Close')

# Step 4: Calculate RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test_actual, rnn_predictions_actual))
print(f'RNN RMSE: {rmse}')

# Visualize the results
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
plt.plot(y_test_actual, label='Actual Price', color='blue')
plt.plot(rnn_predictions_actual, label='RNN Predicted Price', color='red')
plt.title('RNN Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
