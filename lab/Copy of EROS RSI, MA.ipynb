```python
import pandas as pd
import numpy as np

# Function to calculate RSI
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()  # Calculate the difference between consecutive close prices
    gain = np.where(delta > 0, delta, 0)  # Take only the positive differences (gains)
    loss = np.where(delta < 0, -delta, 0)  # Take only the negative differences (losses)

    avg_gain = pd.Series(gain).rolling(window=window, min_periods=1).mean()  # Rolling average of gains
    avg_loss = pd.Series(loss).rolling(window=window, min_periods=1).mean()  # Rolling average of losses

    rs = avg_gain / avg_loss  # Relative Strength (RS)
    rsi = 100 - (100 / (1 + rs))  # RSI formula

    data['RSI'] = rsi  # Add RSI to the dataframe
    return data

# Apply RSI to the Nifty 50 OHLC data
nifty_data = pd.read_csv('/content/NSE50.csv')  # Load data (assumed already saved)
nifty_data = calculate_rsi(nifty_data)

# Display the data with RSI
print(nifty_data[['Date','Open','High','Low','Close','RSI']].tail())

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('/content/NSE50.csv')

# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculate Simple Moving Average (SMA)
def calculate_sma(data, window):
    return data['Close'].rolling(window=window).mean()

# Calculate Exponential Moving Average (EMA)
def calculate_ema(data, window):
    return data['Close'].ewm(span=window, adjust=False).mean()

# Define window sizes for SMA and EMA
windows = [20, 50, 100, 150, 200]

# Calculate SMA and EMA for each window size and add to DataFrame
for window in windows:
    data[f'SMA_{window}'] = calculate_sma(data, window)
    data[f'EMA_{window}'] = calculate_ema(data, window)

# Display the updated DataFrame with SMAs and EMAs
print(data[['Date', 'Close'] + [f'SMA_{window}' for window in windows] + [f'EMA_{window}' for window in windows]].head())

# Optional: Plotting
plt.figure(figsize=(14, 7))
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')

# Plot SMA and EMA
for window in windows:
    plt.plot(data['Date'], data[f'SMA_{window}'], label=f'{window}-Day SMA', linestyle='--')
    plt.plot(data['Date'], data[f'EMA_{window}'], label=f'{window}-Day EMA', linestyle=':')

# Save the DataFrame with SMA and EMA data to a new CSV file
data.to_csv('NSE501.csv', index=False)

plt.title('Nifty 50 Close Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load your data
data = pd.read_csv('/content/NSE50.csv')

# Ensure the 'Date' column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculate RSI (e.g., 14 days)
def calculate_rsi(data, period=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

data['RSI'] = calculate_rsi(data)

# Define window sizes for SMA and EMA
windows = [20, 50, 100, 150, 200]

# Calculate SMA and EMA
for window in windows:
    data[f'SMA_{window}'] = data['Close'].rolling(window=window).mean()
    data[f'EMA_{window}'] = data['Close'].ewm(span=window, adjust=False).mean()

# Drop rows with NaN values
data = data.dropna()

# Select features and target
features = data[['RSI'] + [f'SMA_{window}' for window in windows] + [f'EMA_{window}' for window in windows]]
target = data['Close']

# Normalize the features
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

# Convert target to numpy array
target_scaled = target.values
```

```python
# Define the time step
time_step = 60  # Number of previous days to consider for prediction

# Create the datasets
X, y = [], []
for i in range(time_step, len(features_scaled)):
    X.append(features_scaled[i-time_step:i])
    y.append(target_scaled[i])

X, y = np.array(X), np.array(y)

# Split into training and testing sets (80% training, 20% testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
# Make predictions
predictions = model.predict(X_test)

# Inverse transform the predictions and actual values
predictions = scaler.inverse_transform(predictions)
y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plotting the results
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
plt.plot(data['Date'].iloc[train_size + time_step:], y_test_inverse, label='Actual Price', color='blue')
plt.plot(data['Date'].iloc[train_size + time_step:], predictions, label='Predicted Price', color='orange')
plt.title('Nifty 50 Price Prediction with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Load your dataset
data = pd.read_csv('/content/NSE501.csv')
data['Date'] = pd.to_datetime(data['Date'])

# Calculate moving averages
data['MA20'] = data['Close'].rolling(window=20).mean()
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA100'] = data['Close'].rolling(window=100).mean()
data['MA150'] = data['Close'].rolling(window=150).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()

# Drop rows with NaN values
data = data.dropna()

# Select features and target
features = data[['MA20', 'MA50', 'MA100', 'MA150', 'MA200']]
target = data['Close']

# Normalize features and target
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

features_scaled = feature_scaler.fit_transform(features)
target_scaled = target.values.reshape(-1, 1)
target_scaled = target_scaler.fit_transform(target_scaled)

# Create datasets for RNN
time_step = 10
X, y = [], []
for i in range(time_step, len(features_scaled)):
    X.append(features_scaled[i-time_step:i])
    y.append(target_scaled[i])

X, y = np.array(X), np.array(y)

# Split into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the RNN model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
rnn_model.add(Dropout(0.2))
rnn_model.add(SimpleRNN(50, return_sequences=False))
rnn_model.add(Dropout(0.2))
rnn_model.add(Dense(1))

# Compile the model
rnn_model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = rnn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, callbacks=[early_stopping])

# Make predictions
rnn_predictions = rnn_model.predict(X_test)

# Inverse transform the predictions and actual values
rnn_predictions_inverse = target_scaler.inverse_transform(rnn_predictions)
y_test_inverse = target_scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluate model performance
mse = mean_squared_error(y_test_inverse, rnn_predictions_inverse)
print(f'Mean Squared Error for RNN: {mse}')

# Plotting the results
plt.figure(figsize=(14, 7))
plt.plot(data['Date'].iloc[train_size + time_step:], y_test_inverse, label='Actual Price', color='blue')
plt.plot(data['Date'].iloc[train_size + time_step:], rnn_predictions_inverse, label='Predicted Price', color='orange')
plt.title('Nifty 50 Price Prediction with RNN Using Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
from kerastuner import RandomSearch

def build_model(hp):
    model = Sequential()
    model.add(SimpleRNN(hp.Int('units', min_value=50, max_value=100, step=10), return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.2))
    model.add(SimpleRNN(hp.Int('units', min_value=50, max_value=100, step=10), return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

tuner = RandomSearch(build_model, objective='val_loss', max_trials=5, executions_per_trial=1)

tuner.search(X_train, y_train, epochs=50, validation_split=0.1)
best_model = tuner.get_best_models(num_models=1)[0]
```
